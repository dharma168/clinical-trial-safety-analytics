```python
# Create Manual Mapping
dm["SEX_LABEL"] = dm["SEX"].map({1.0: "Male"})
```

```python
race_map = {
    0.0: "White",
    1.0: "Black",
    2.0: "Hispanic or Latino",
    3.0: "Asian",
    4.0: "American Indian or Alaskan Native",
    5.0: "Native Hawaiian or Other Pacific Islander",
    99.0: "Other"
}

dm["RACE_LABEL"] = dm["RACE"].map(race_map)
```

```python
dm["CONSENT_LABEL"] = dm["CNSNT"].map({
    1.0: "Consented",
    0.0: "Not Consented"
})
```
